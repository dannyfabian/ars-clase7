## EJEMPLO DE GRAFOS CON NETWORX


```python
import networkx as nx
```

```python
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
G= nx.Graph()
G.add_edge('1', '2', weight=10)
G.add_edge('1', '4', weight=9)
G.add_edge('2', '4', weight=6)
G.add_edge('2', '3', weight=5)
G.add_edge('4', '5', weight=5)
G.add_edge('5', '6', weight=6)
G.add_edge('3', '6', weight=9)
G.add_edge('3', '5', weight=7)
```

```python
pos= nx.spring_layout(G)
nx.draw_networkx_edges(G, pos, node_sizes=300)
nx.draw_networkx_edges(G, pos, width=6)
nx.draw_networkx_labels(G, pos, font_size=10, font_family = 'sans-serif')
nx.draw_networkx_edge_labels(G, pos)
plt.axis('off')
plt.show()
```

```python
nx.density(G)
```

```python
nx.degree_histogram(G)
```

```python
for p in nx.all_neighbors(G,'1'):
    print(p)
```

```python
nx.degree_centrality(G)
```

```python
nx.number_of_nodes(G)
```

```python
nx.clustering(G)
```

```python
A=nx.adjacency_matrix(G, pos, 64)
```

```python
A.toarray()
```

```python
B=nx.laplacian_matrix(G, pos, 6)
B.toarray()
```
